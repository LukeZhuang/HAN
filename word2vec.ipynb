```python
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
year='2013'

f_train=open('data/yelp'+year+'/yelp-'+year+'-train-nonvocab.txt')
# f_val=open('data/yelp'+year+'/yelp-'+year+'-dev-nonvocab.txt')
fo=open('data/yelp'+year+'/mixed.txt','w')

for line in f_train:
    document=line.strip().split('\t\t')[1]
    for s in document.split('\t'):
        fo.write(s+'\n')
        
# for line in f_val:
#     document=line.strip().split('\t\t')[1]
#     for s in document.split('\t'):
#         fo.write(s+'\n')
        
f_train.close()
# f_val.close()
fo.close()
```

```python
fmixed=open('data/yelp'+year+'/mixed.txt')
```

```python
output1 = 'data/yelp'+year+'/yelp_'+year+'_50.vector'
output2 = 'data/yelp'+year+'/yelp_'+year+'_50.vocab'

model = Word2Vec(LineSentence(fmixed), size=50, window=5, min_count=5, workers=4)
model.wv.save_word2vec_format(output1, output2, binary=False)
```

```python
print(model.most_similar('ridiculous'))
print(model['ridiculous'])
```
