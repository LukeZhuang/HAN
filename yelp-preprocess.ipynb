```python
import numpy as np
import codecs
import re
# from nltk.corpus import stopwords
# stop_words = set(stopwords.words('english'))
```

```python
spaces=re.compile('\s+')
nonalpnum=re.compile('[^0-9a-zA-Z]+')
multi_empty_sent=re.compile(' (\t )+')

def process(sentence):
    words=spaces.split(sentence)
    new_words=[]
    for word in words:
#         if word in stop_words:
#             continue
        if word=='-lrb-' or word=='-rrb-':
            continue
        if word=='<sssss>':
            new_words.append('\t')
            continue
        tt=nonalpnum.split(word)
        t=''.join(tt)
        if t!='':
            new_words.append(t)
    new_sentence=' '.join(new_words)
    new_sentence=re.sub(pattern='( \t)+ *', repl='\t', string=new_sentence)
    if new_sentence!='' and new_sentence[-1]=='\t':
        new_sentence=new_sentence[:-1]
    return new_sentence
```

```python
year=['2013','2014','2015']
dataset=['train','dev','test']
# year=['2013']
# dataset=['dev']

for y in year:
    for d in dataset:
        print(y,d)
        f=codecs.open('data/yelp'+y+'/yelp-'+y+'-'+d+'-Copy1.txt.ss','r', encoding='utf-8')
        fo=open('data/yelp'+y+'/yelp-'+y+'-'+d+'-nonvocab.txt','w')
        cnt=0
        for line in f:
            cols=line.strip().split('\t\t')
            sentence=cols[3]
            rate=cols[2]
            new_sentence=process(sentence)
            if new_sentence!='':
                fo.write(rate+'\t\t'+new_sentence+'\n')
            cnt+=1
            if cnt==0 or (cnt+1)%10000==0:
                print('finish',cnt+1,'sentences')
        print(cnt)
        f.close()
        fo.close()
```

```python
year=['2013','2014','2015']
for y in year:
    print(y)
    f=open('data/yelp'+y+'/yelp-'+y+'-train-nonvocab.txt')
    vocab_file=open('data/yelp'+y+'/vocab-'+y+'.txt','w')
    freq={}
    cnt=0
    for line in f:
        document=line.strip().split('\t\t')[1]
        sents=document.split('\t')
        for s in sents:
            words=s.split(' ')
            for w in words:
                if w in freq.keys():
                    freq[w]+=1
                else:
                    freq[w]=1
        cnt+=1
        if cnt==0 or (cnt+1)%50000==0:
            print('finish',cnt+1,'sentences')
    freq_items=sorted(freq.items(), key=lambda t: t[1], reverse=True)
    for (i,v) in enumerate(freq_items):
        vocab_file.write(v[0]+'\t'+str(v[1])+'\n')
        if (i+1)==50000:
            break
    f.close()
    vocab_file.close()
```
